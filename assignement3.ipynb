```python
import pandas as pd

# Loads the dataset
df = pd.read_csv("kiva_loans.csv")
display(df)
```

```python
# Selects only the food sector
food_sector_df = df[df["sector"].isin(["Food",])]
display(food_sector_df)
```

```python
# Converts the currencies to euro
from currency_converter import CurrencyConverter
c = CurrencyConverter()

# List of supported currencies in CurrencyConverter
supported_currencies = c.currencies
food_sector_df = food_sector_df[food_sector_df["currency"].isin(supported_currencies)]

# Apply conversion
food_sector_df["loan_amount_euro"] = food_sector_df.apply(lambda row: c.convert(row["loan_amount"], row["currency"], "EUR"), axis=1)

# Display converted DataFrame
display(food_sector_df)
```

```python
# Sum of food sector loan amount grouped by country and acitivity
food_sector_df = food_sector_df.groupby(["country","activity"]).agg({        
    "loan_amount_euro": "sum",              
}).reset_index()

# Rename the column of loan amount to add th prefix total
food_sector_df = food_sector_df.rename(columns={
    "loan_amount_euro": "total_loan_amount",
})

display(food_sector_df)
```

```python
# Sort the total loan amount by descending order
food_sector_df = food_sector_df.sort_values(by="total_loan_amount", ascending = False)

food_sector_df['total_loan_amount'] = food_sector_df['total_loan_amount'].round(0)
display(food_sector_df)
```

```python
# Exludes where the total amount is zero
food_sector_df = food_sector_df[food_sector_df["total_loan_amount"] > 0]
display(food_sector_df)
```

```python
# Dataframe that filters where the total loan amount is greater that the mean per that activity
higher_than_mean_df = food_sector_df[food_sector_df.apply(
    lambda row: row['total_loan_amount'] > food_sector_df[food_sector_df['activity'] == row['activity']]['total_loan_amount'].mean(),
    axis=1
)]

display(higher_than_mean_df)
```
